## Drawing Line Plots

```python
## Line Plotting

from matplotlib import pyplot as plt
import numpy as np

x=np.arange(1,10,0.1)
y=2*x+5

plt.plot(x,y)
plt.show()
```

```python
x=np.arange(1,10,0.1)
y=2*x+5
plt.plot(x,y,color='r')
plt.title('Linear Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

### Drawing Sub-Line Plots

```python
x=np.arange(1,10,0.1)
y=2*x+5
z=3*x+10

plt.subplot(1,2,1)
plt.plot(x,y)


plt.subplot(1,2,2)
plt.plot(x,z)
plt.show()
```

## Bar Graph Plotting

```python
# Bar Plotting

from matplotlib import pyplot as plt
import numpy as np

fruits = {'Banana':30,'Orange':50,'Mango':70,'Apple':100}
names = list(fruits.keys())
price = list(fruits.values())
names,price
```

```python
plt.bar(names,price)
plt.show()
```

```python
plt.bar(names,price)
plt.title('Market prices of Fruits')    # Adds Title For Plot
plt.xlabel('Friuts')                    # Labels the graph axis
plt.ylabel('Prices')
plt.show()
```

```python

plt.bar(names,price,color='yellow')    # Uses to Change the color of the plot
plt.title('Market prices of Fruits')
plt.xlabel('Friuts')
plt.ylabel('Prices')
plt.show()
```

```python
plt.bar(names,price,color='orange')
plt.title('Market prices of Fruits')
plt.xlabel('Friuts')
plt.ylabel('Prices')
plt.show()
```

## Horizontal Bar Plotting

```python
# horizontal Plotting
plt.barh(names,price,color='green')
plt.title('Market prices of Fruits')
plt.xlabel('Friuts')
plt.ylabel('Prices')
plt.show()
```

## Scatter Plotting

```python

# Scatter plot

x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,9,8,2,1]
b = [5,8,5,9,3,4,5,2,4]
plt.scatter(x,a)
plt.scatter(x,b)
plt.show()

```

```python

x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,9,8,2,1]
b = [5,8,5,9,3,4,5,2,4]
plt.scatter(x,a)
plt.scatter(x,b)
plt.legend(['a','b'])     # Adds the label on the grapg with representation
plt.show()
```

```python

x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,9,8,2,1]
b = [5,8,5,9,3,4,5,2,4]
plt.scatter(x,a)
plt.scatter(x,b)
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.xlabel('X')
plt.ylabel('a&b')
plt.show()
```

```python

x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,9,8,2,1]
b = [5,8,5,9,3,4,5,2,4]
plt.scatter(x,a,s=200)    # Changing the size of the bubble
plt.scatter(x,b,s=200)
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.xlabel('X')
plt.ylabel('a&b')
plt.show()
```

```python

x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,9,8,2,1]
b = [5,8,5,9,3,4,5,2,4]
plt.scatter(x,a,s=200,marker='3')      # Changing the Shape of the Bubble
plt.scatter(x,b,s=200,marker='1')
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.xlabel('X')
plt.ylabel('a&b')
plt.show()
```

```python

x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,9,8,2,1]
b = [5,8,5,9,3,4,5,2,4]

plt.figure(figsize=(10,5))    # Setting Figure Size as required
plt.scatter(x,a,s=200)
plt.scatter(x,b,s=200)
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.xlabel('X')
plt.ylabel('a&b')
plt.show()
```

```python

x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,9,8,2,1]
b = [5,8,5,9,3,4,5,2,4]

plt.figure(figsize=(10,10))
plt.scatter(x,a,s=200)
plt.scatter(x,b,s=200)
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.xlabel('X')
plt.ylabel('a&b')
plt.show()
```

## Pi Charts

```python
# Pi Charts

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

plt.pie(price,labels=fruits)
plt.show()
```

### Dividing the Persentage to the pi plot

```python

fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

plt.pie(price,labels=fruits,autopct='%0.1f%%')       # Uses for representating the persentage of the space occupied
plt.show()
```

```python

fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

plt.pie(price,labels=fruits,autopct='%0.1f%%',shadow=True)    # Shows The Shadow
plt.show()
```

### Slicing the Pi Chart using its indexing values based on '0' & '1' 

```python

fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

plt.pie(price,labels=fruits,autopct='%0.1f%%',explode=(0,0,0.1,0,0))      # Slicing the required area from the pi plot
plt.show()
```

```python

fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

plt.pie(price,labels=fruits,autopct='%0.1f%%',explode=(0.2,0,0.2,0,0))
plt.show()
```

### Keeping Our required colors in pi chart

```python

fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

plt.pie(price,labels=fruits,autopct='%0.1f%%',shadow=True,colors=['black','blue','green','yellow','red'])
plt.show()
```

## Doughnut Plot

```python
#Doughnut_Charts

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

pie1 = plt.pie(price,labels=fruits,autopct='%0.1f%%')
pie2 = plt.pie([5],colors='w')
plt.show()
```

```python

fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

pie1 = plt.pie(price,labels=fruits,autopct='%0.1f%%',radius=2)
pie2 = plt.pie([5],colors='w',radius=1)
plt.show()
```

```python

fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

pie1 = plt.pie(price,labels=fruits,autopct='%0.1f%%',radius=2)
pie2 = plt.pie([5],colors='w',radius=0.5)
plt.show()

```

```python

fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

pie1 = plt.pie(price,labels=fruits,autopct='%0.1f%%',radius=2)
pie2 = plt.pie([5],colors='w',radius=1)
pie3 = plt.pie([5],colors='y',radius=0.5)
plt.show()
```

## Box Plot

```python
# Box Plot

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

x = [1,2,3,4,5,6,7,8,9]
y = [1,2,3,4,5,4,3,2,1]
z = [6,7,8,9,8,7,6,5,4]

data = list([x,y,z])

plt.boxplot(data)
plt.show()

```

## Violin Plot

```python

x = [1,2,3,4,5,6,7,8,9]
y = [1,2,3,4,5,4,3,2,1]
z = [6,7,8,9,8,7,6,5,4]

data = list([x,y,z])

#plt.boxplot(data)
plt.violinplot(data)
plt.show()
```

```python

x = [1,2,3,4,5,6,7,8,9]
y = [1,2,3,4,5,4,3,2,1]
z = [6,7,8,9,8,7,6,5,4]

data = list([x,y,z])

plt.violinplot(data)
plt.grid(True)            # Adding Grids to plot
plt.show()
```

```python

x = [1,2,3,4,5,6,7,8,9]
y = [1,2,3,4,5,4,3,2,1]
z = [6,7,8,9,8,7,6,5,4]

data = list([x,y,z])

plt.violinplot(data,showmedians=True)     # Function for showing the Medians Values of the graph
plt.title('Distribution of Data')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python

x = [1,2,3,4,5,6,7,8,9]
y = [1,2,3,4,5,4,3,2,1]
z = [6,7,8,9,8,7,6,5,4]

data = list([x,y,z])

plt.violinplot(data,showmedians=True,showmeans=True)     # Function for showing the Mean Values of the graph
plt.title('Distribution of Data')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```
